```python
import csv
import matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import NeuralNetwork as nn
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn import metrics
from sklearn.cross_validation import cross_val_score
```

```python
def read_files():
    global hero, test, train, sample
    hero = pd.read_csv('Data/MatchDetail.csv')
    test = pd.read_csv('Data/MatchOverviewTest.csv')
    train = pd.read_csv('Data/MatchOverviewTraining.csv')
    sample = pd.read_csv('Data/sampleSubmission.csv')
    train.index = train.match_id.tolist()
    del train['match_id']
read_files()
train[(24<train)&(train<108)] -= 1
train[(108 < train) & (train < 114)] -= 2
train.loc[:,'hero_6':'hero_10'] += 111
length = len(train)
X = pd.DataFrame({i:[0 for _ in range(length)] for i in range(1, 223)}, index = train.index)
for index, row in train.iterrows():
    for hero_id in row.tolist()[:10]:
        X.loc[index, hero_id] = 1
y = train.first_5_won.astype('int')
```

```python
# Creates copies of the datasets
xcopy = X.copy()
ycopy = y.copy()
```

```python
# Reinitializes data to avoid having to reread files
X = xcopy.copy()
y = ycopy.copy()
```

```python
# Only converts to np.array if it isn't already
import NeuralNetwork as nn
try:
    y = np.array([y.values]).T
    X = X.values
except:
    pass
X = X - 0.5
X = nn.add_bias(X)
dota_nn = nn.NeuralNetwork((223, 100, 1), momentum=.5, learning_rate=1.5)
```

```python
dota_nn.train(X, y, num_loops = 500)
```
